```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
#loading data from sklearn
breast_cancer_detection = sklearn.datasets.load_breast_cancer()
```

```python
#loading the data in a dataframe

df = pd.DataFrame(breast_cancer_detection.data, columns = breast_cancer_detection.feature_names)
```

```python
#seperating the features and target
X = df.drop(columns = 'label', axis = 1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training and testing data
X_train,X_test,Y_train,Y_test  = train_test_split(X,Y, test_size = 0.2,random_state = 2) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#standardize the data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.fit_transform(X_test)
```

```python
print(X_train_std)
```

```python
pip install tensorflow
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of Neural Network
model = keras.Sequential([
                            keras.layers.Flatten(input_shape = (30,)),
                            keras.layers.Dense(20, activation= 'relu'),
                            keras.layers.Dense(2, activation= 'sigmoid')
])
```

```python
# Compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, validation_split = 0.1, epochs = 10)
```

```python
#visualizing accuracy and loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')

plt.legend(['training_data','validation_data'],loc = 'lower right')
```

```python
#visualizing accuracy and loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training_data','validation_data'],loc = 'upper right')
```

```python
#Accuracy of the model on test data

loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
